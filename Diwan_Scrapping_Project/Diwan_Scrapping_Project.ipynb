### Import Helper Libraries 

```python
import csv,requests,bs4,lxml
from bs4 import BeautifulSoup
```

### Request and get HTML page content and parse it 

```python
url = 'https://diwanegypt.com/product-category/books/english-books/'
newurl= 'https://diwanegypt.com/product-category/books/english-books/page/1/'
print(newurl)
```

```python
req = requests.get(newurl)
soup= BeautifulSoup(req.content,'lxml')
```

### Get One book data 
* name 
* author 
* price 

```python
sectionData = soup.find('div',{'class':'section-inner products-overview-with-sidebar'}).find('div',{'class':'inner'})
ListOFli = sectionData.find('ul',{'class':'products columns-5'}).find_all('li')
temp = ListOFli[0].find('a',{'class':'woocommerce-LoopProduct-link woocommerce-loop-product__link'})
name = temp.find('h2').text.strip()
author = temp.find('span',{'class':'author'}).text.strip()
price = temp.find('span',{'class':'price'}).find('bdi').text.strip().removeprefix('EGP').replace(',','.')
print(name,' - ',author,' - ',price)
    
```

### Get All the books of the page 

```python
def pageData(soup):
    sectionData = soup.find('div',{'class':'section-inner products-overview-with-sidebar'}).find('div',{'class':'inner'})
    ListOFli = sectionData.find('ul',{'class':'products columns-5'}).find_all('li')
    for li in ListOFli:
        temp = li.find('a',{'class':'woocommerce-LoopProduct-link woocommerce-loop-product__link'})
        name = temp.find('h2').text.strip()
        author = temp.find('span',{'class':'author'}).text.strip()
        price = temp.find('span',{'class':'price'}).find('bdi').text.strip().removeprefix('EGP').replace(',','')
        price = float(price)
        print(name,' - ',author,' - ',price)
```

```python
pageData(soup)
```

### Get All books of All pages 

```python
def allPages():
    for page in range(1,91):
        print(f'page number {page}')
        newurl= f'https://diwanegypt.com/product-category/books/english-books/page/{page}/'
        req = requests.get(newurl)
        soup= BeautifulSoup(req.content,'lxml')
        pageData(soup)
        print('*'*120)
```

```python
allPages()
```
