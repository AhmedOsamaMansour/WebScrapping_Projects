```python
import csv,requests,os,bs4,lxml
from bs4 import BeautifulSoup
```

```python
# list for each page
jobsData = []
def main():
    for i in range(52):
        wazzaf_URL=f'https://wuzzuf.net/search/jobs/?a=navbg&q=data%20science&start={i}'
        print(wazzaf_URL)
        req = requests.get(wazzaf_URL)
        soup = BeautifulSoup(req.content,'lxml')
        onePage(soup.find('div',attrs={'class':'css-9i2afk'}))

def oneJob(box):
    temp = box.find('div',attrs={'class':'css-laomuu'})
    jobName = temp.find('h2',attrs={'class':'css-m604qf'}).text.strip()
    companyName = temp.find('div',attrs={'class':'css-d7j1kk'}).find('a').text[:-1].strip()
    companyLocation = temp.find('div',attrs={'class':'css-d7j1kk'}).find('span').text.strip()
    jobLink = temp.find('h2').find('a')['href']
    print(jobLink)
    jobDesc = box.find('div',attrs={'class':'css-y4udm8'}).find('div',attrs=None).text.strip()
    jobsData.append((jobName,companyName,companyLocation,jobDesc,jobLink))

def storeData(jobs):
    # store Jobs in csv :
    m = 'a' if os.path.exists('wazzaf_DS.csv') else 'w'
    with open('wazzaf_DS.csv',mode=m,newline='') as f:
        w = csv.writer(f)
        if m == 'w':
            w.writerow(['Job Name','Company Name','Company Location','Job Description','Job Link'])
        w.writerows(jobs)

def onePage(page):
    Jobs = page.find_all('div',attrs={'class':'css-1gatmva e1v1l3u10'})
    for job in Jobs:
        oneJob(job)
    #store data in csv
    storeData(jobsData)
    jobsData.clear()

main()
```
